###서울시의 지하철역의 주소를 기준으로 동만 추출

```python
#'서울특별시'로 시작하는 행만 필터링
df_seoul = df[df['구주소'].str.startswith('서울특별시')]

#공백으로 나누어 동 정보 추출 (세 번째 요소)
df_seoul['동'] = df_seoul['구주소'].str.split().str[2]
```

```python
df_seoul
```

```python
df_seoul[df_seoul['역명']=='숙대입구(갈월)'] #잘 생성되었는지 확인
```

```python
df_seoul[df_seoul['역명']=='강일']
```

```python
subway_count = df_seoul.groupby('동').size().reset_index(name='역 개수') #동별로 역의 개수
```

```python
subway_count[subway_count['동']=='상일동']
```

```python
subway_count.to_excel('동별 지하철 개수.xlsx', index=False)
```

###등록인구 데이터와 합치기

```python
#등록인구 데이터
population_df = pd.read_excel("/content/drive/MyDrive/population_data.xlsx")
population_df
```

```python
# 동 이름 기준으로 병합
merged_df = pd.merge(population_df, subway_count, left_on='dong_name', right_on='동', how='left')

# 지하철 개수가 없는 경우 NaN → 0으로 처리
merged_df['역 개수'] = merged_df['역 개수'].fillna(0).astype(int)

merged_df = merged_df.drop(columns=['동'])

merged_df.rename(columns={'역 개수': 'subway_count'}, inplace=True)

merged_df
```

```python
merged_df.to_excel('등록인구 기준 동별 지하철 개수.xlsx', index=False)
```
