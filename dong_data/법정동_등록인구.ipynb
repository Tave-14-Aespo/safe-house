```python
dong_id = dong_id.drop(columns=['생성일자','말소일자'])
dong_id = dong_id.dropna()
dong_id.rename(columns={'읍면동명': '행정동', '동리명': '법정동'}, inplace=True)
```

```python
import requests
import xml.etree.ElementTree as ET
import pandas as pd

# API 엔드포인트 및 키
url = 'http://apis.data.go.kr/1741000/stdgPpltnHhStus/selectStdgPpltnHhStus'
api_key = 'Jwx7E8hpKhLhxJwKlAWtUUovu2QBvwOafGMEI6FHSjXPm+B4MZrxdNLCU8myo8hbpEHfA6oHC3WKfqvaUBbpNw=='

# 결과를 저장할 리스트
results = []

# 각 구 코드를 순회하며 데이터 요청
for _, row in gu_id.iterrows():
    gu_name = row['시군구명']
    gu_code = row['법정동코드']

    # 요청 파라미터
    params = {
        "serviceKey": api_key,
        "stdgCd": gu_code,
        "srchFrYm": "202411",
        "srchToYm": "202411",
        "lv": "3",
        "regSeCd": "1",
        "type": "XML",
        "numOfRows": 100,
        "pageNo": 1
    }

    # API 요청
    response = requests.get(url, params=params)

    if response.status_code == 200:
        try:
            # XML 데이터 파싱
            root = ET.fromstring(response.text)
            for item in root.findall('.//item'):
                # 필요한 데이터 추출
                stdgNm = item.find('stdgNm').text  # 동 이름
                stdgCd = item.find('stdgCd').text  #법정동 코드
                totNmprCnt = item.find('totNmprCnt').text  # 인구 수

                # 결과 리스트에 추가
                results.append({
                    'gu_name': gu_name,  # 구 이름
                    'dong_name': stdgNm,
                    'dong_code': int(stdgCd),
                    'population': int(totNmprCnt)
                })
        except Exception as e:
            print(f"{gu_name} 데이터 파싱 에러:", e)
    else:
        print(f"{gu_name} API 호출 실패: {response.status_code}")
        print(response.text)

# 데이터프레임 생성
population_df = pd.DataFrame(results)

# 데이터프레임 출력
print(population_df)
```

```python
# population_df에서 법정동코드 추출
population_codes = population_df['dong_code'].unique()

# dong_id에서 법정동코드 추출
dong_id_codes = dong_id['법정동코드'].unique()

# dong_id에는 있지만 population_df에는 없는 값 찾기
missing_codes = set(dong_id_codes) - set(population_codes)

# 결과 출력
print("population_df에 없는 법정동코드:")
print(missing_codes)

print("\n")

# dong_id에서 missing_codes에 해당하는 데이터 추출
missing_df = dong_id[dong_id['법정동코드'].isin(missing_codes)]
```
