```python
#필요 라이브러리 임포트
import numpy as np
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
```

```python
#데이터 불러오기
import pandas as pd
data = pd.read_csv('/content/drive/MyDrive/오피스텔_최종.csv', encoding = 'cp949')
```

```python
data.head()
```

```python
#데이터 스케일링
scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(data[['bus_count', 'cctv_count', 'school_count', 'light_count']])
X_scaled_df = pd.DataFrame(X_scaled, columns=['bus_count_scaled', 'cctv_count_scaled', 'school_count_scaled', 'light_count_scaled'])
```

```python
# y 값 생성
# y = bus_count + cctv_count + school_count + light_count (가중치 모두 1)
weights = np.array([1, 1, 1, 1])  # 가중치
y = np.dot(X_scaled, weights)  # y 계산
y = (y / y.max()) *100 # y 값을 0 ~ 100 사이로 스케일링
data['y'] = y
```

```python
#다중 회귀 모델 구현 (절편 0)
# 수식: β = (X^T X)^(-1) X^T y
weights_estimated = np.linalg.inv(X_scaled.T @ X_scaled) @ X_scaled.T @ y
```

```python
#예측값 계산
y_pred = np.dot(X_scaled, weights_estimated)
print(y_pred)
```

```python
#스케일링된 값과 y 예측값 추가
data[['bus_count_scaled', 'cctv_count_scaled', 'school_count_scaled', 'light_count_scaled']] = X_scaled_df
data['y_pred'] = y_pred
print(data)
```

```python
#동만 출력하여 새로운 열에 저장
data['동'] = data['시군구'].str.replace(r'서울특별시\s[가-힣]+구\s', '', regex=True)
```

```python
print(data)
```
